```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load the Titanic dataset
titanic = pd.read_csv('titanic.csv')

# Create a contingency table (cross-tabulation) for Pclass and Survived
ct=pd.crosstab(titanic['Pclass'], titanic['Survived'])

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(ct,annot=True,fmt='d')
plt.title('Heatmap of Passenger Class vs Survival')
plt.xlabel('Survived')
plt.ylabel('Passenger Class')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset
titanic = pd.read_csv('titanic.csv')

# Select only numerical columns for correlation analysis
numerical_features = titanic.select_dtypes(include=['float64', 'int64'])

# Compute the correlation matrix
correlation_matrix = numerical_features.corr()

# Set up the matplotlib figure
plt.figure(figsize=(10, 8))

# Generate a heatmap
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, cbar_kws={"shrink": .8})

# Add titles and labels
plt.title('Correlation Matrix of Titanic Dataset Features')
plt.show()
```

```python
## KNN Imputer

import pandas as pd
import seaborn as sns
from sklearn.impute import KNNImputer

# Load the Titanic dataset
titanic = pd.read_csv('titanic.csv')

# Display missing values before imputation
print("Missing values before imputation:\n", titanic.isnull().sum())

# Select relevant columns for KNN imputation
# We'll use 'Age' and 'Fare' as numerical features for the imputer
# We can also convert categorical columns to numeric
titanic['Sex'] = titanic['Sex'].map({'male': 0, 'female': 1})  # Convert 'sex' to numerical
titanic['Embarked'] = titanic['Embarked'].map({'C': 0, 'Q': 1, 'S': 2})  # Convert 'embarked' to numerical

# Select columns for KNN imputation
knn_columns = ['Age', 'Fare', 'Sex', 'Embarked']
titanic_knn = titanic[knn_columns]

# Initialize KNN Imputer
imputer = KNNImputer(n_neighbors=5)

# Fit and transform the data using KNN imputer
titanic_imputed = imputer.fit_transform(titanic_knn)

# Create a DataFrame with the imputed data
imputed_df = pd.DataFrame(titanic_imputed, columns=knn_columns)

# Display missing values after imputation
print("\nMissing values after imputation:\n", imputed_df.isnull().sum())

# Display a sample of the imputed dataset
print("\nSample of the imputed dataset:\n", imputed_df.head(10))
```

```python
### One-Hot Encoding

import pandas as pd
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder

# Load the Titanic dataset
titanic = pd.read_csv('titanic.csv')

# Display original 'sex' and 'embarked' columns
print("Original 'sex' and 'embarked' columns:\n", titanic[['Sex','Embarked']].head())

# Apply One-Hot Encoding on 'sex' and 'embarked' columns
# Use pandas get_dummies for simplicity
titanic_onehot = pd.get_dummies(titanic, columns=['Sex','Embarked'],dtype=int)

# Display the one-hot encoded columns
print("\nOne-Hot Encoded Titanic dataset (first few rows):\n", titanic_onehot.head())
```

**Feature Scaling**

```python
### Min_Max Scaler

import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Step 1: Load the Titanic dataset
titanic = pd.read_csv('titanic.csv')

# Step 2: Select features for scaling
# We will select 'Age' and 'Fare' for scaling
features_to_scale = ['Age']

# Step 3: Handle missing values
# Fill missing values in 'Age' and 'Fare' with the median
titanic['Age'].fillna(titanic['Age'].median(), inplace=True)

# Step 4: Apply MinMaxScaler to the entire dataset
scaler = MinMaxScaler()

# Fit the scaler on the entire dataset and transform it
titanic[features_to_scale] = scaler.fit_transform(titanic[features_to_scale])

# Step 5: Display the results
print("After MinMax Scaling:")
print(titanic[features_to_scale].head())
```

# **Conclusion:**


*   Use MinMaxScaler when you want to normalize data between a fixed range, and especially when working with distance-based algorithms or models that are sensitive to feature magnitudes.

**Feature Creation**

```python
#### Feature Creation

import pandas as pd

# Step 1: Load the Titanic dataset
titanic = pd.read_csv('titanic.csv')

# Step 2: Inspect the dataset to understand the relevant columns
# SibSp = Number of Siblings/Spouses aboard
# Parch = Number of Parents/Children aboard

# Step 3: Create a new feature 'FamilySize'
titanic['FamilySize'] = titanic['SibSp'] + titanic['Parch'] + 1

# Step 4: Display the results
print(titanic[['SibSp', 'Parch', 'FamilySize']].head())
```

# **Conclusion:**
The Family Size feature could be a good predictor in the Titanic dataset. For example:


*   Small families (e.g., FamilySize = 1, traveling alone) may have different survival rates compared to large families.
*   It allows us to capture a more nuanced relationship between family members and survival probability.

**Feature Selection**

```python
#### Feature Selection

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Load the Titanic dataset
titanic = pd.read_csv('titanic.csv')

# Step 2: Select numerical features for correlation analysis
# We can select numerical columns like 'Age', 'Fare', 'SibSp', 'Parch'
numerical_features = ['Survived', 'Age', 'Fare', 'SibSp', 'Parch']

# Step 3: Handle missing values
# We'll fill missing values in the 'Age' feature with the median
titanic['Age'].fillna(titanic['Age'].median(), inplace=True)

# Step 4: Compute the correlation matrix
corr_matrix = titanic[numerical_features].corr()

# Step 5: Visualize the correlation matrix with a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt='.2f')
plt.title("Correlation Matrix of Titanic Dataset")
plt.show()
```

# **Conclusion:**


*   Features that are highly correlated with each other or show weak correlation with the target variable might be removed, or they can be combined or transformed.
*   The correlation matrix helps identify these relationships and informs decisions about feature selection for model training.
